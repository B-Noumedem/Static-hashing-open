```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=2)
import seaborn as sns
import glob
sns.set_style("whitegrid")
```

```python
txt_files = glob.glob('*.txt')
times = {}
for text_file_path in txt_files:
    text_file = open(text_file_path, "r")
    lines = text_file.readlines()[0]
    lines = [int(x) for x in lines[1:(len(lines)-2)].split(',')]
    times[text_file_path[0:-4]]= lines
    text_file.close()
```

```python
y = times['bloc_Insertions_times_ST'] 
nb_fold = len(y)
step = 50000
x = np.array(range(0,nb_fold*step, step))/1e6
plt.figure(figsize=(8,5))
sns.barplot(x=x, y=y, color='blue')
plt.xlabel('Hashtable size before insertion (*1e6)')
plt.ylabel('Time execution (nano)')
plt.title('Singlethreading Insertion performance',fontsize=16)
```

```python
y = times['bloc_Insertions_times_MT']
nb_fold = len(y)
step = 50000
x = np.array(range(0,nb_fold*step, step))/1e6
plt.figure(figsize=(8,5))
sns.barplot(x=x, y=y, color='blue')
plt.xlabel('Hashtable size before insertion')
plt.ylabel('Time execution (nano)')
plt.title('Multithreading Insertion performance',fontsize=16)
```

```python
#x = np.arange(len(y))
nb_fold = len(times['bloc_Insertions_times_ST'])
nb_search = 1000
y = np.array(times['bloc_Search_times_ST']).reshape((nb_fold, nb_search)).mean(axis=1)
x = np.array(range(1,nb_fold+1))
plt.figure(figsize=(8,5))
sns.barplot(x=x, y=y, color='blue')
plt.xlabel('Batch number')
plt.ylabel('Time execution (nano)')
plt.title('Singlethreading search performance',fontsize=16)
```
